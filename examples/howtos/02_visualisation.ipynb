# Visualisation

A short tour of how to *see* a scenario: the plan (geometry, exits, journeys), an interactive replay of a run, and the two overlaid.

```python
import jupedsim_scenarios as js
```

## Resolve the bundled scenario

We ship example scenarios under `examples/assets/`. This resolves the path whether you run the notebook from the repo root, `examples/`, or `examples/howtos/`.

```python
from pathlib import Path

scenario_path = next(
    p
    for p in [
        Path("examples/assets/quickstart.zip"),
        Path("../assets/quickstart.zip"),
        Path("assets/quickstart.zip"),
    ]
    if p.exists()
)
```

## Inspect the plan

`scenario.plot()` draws the walkable area with labelled distributions, exits, zones, and checkpoints. `show_journeys=True` adds routing arrows when the scenario defines journeys.

```python
scenario = js.load_scenario(scenario_path)
scenario.plot(show_journeys=True)
```

## Run the simulation

```python
result = js.run_scenario(scenario=scenario)
```

## Interactive playback

`result.visualise()` returns a plotly figure: agents coloured by speed, with a play button and a time slider. It renders inline — just make it the last line of the cell.

```python
result.visualise()
```

## Plan vs. reality

Pass `trajectories=result` to overlay the agent paths on the geometry — one figure showing the plan and what the agents actually did.

```python
scenario.plot(show_journeys=True, trajectories=result)
```

## Save the animation

Write a self-contained, interactive HTML file you can open in any browser.

```python
result.visualise(save_path="run.html")
```
