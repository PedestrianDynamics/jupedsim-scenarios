# How do I set a waiting time on a checkpoint?

A **checkpoint** is a stage that agents pass through on their journey.
Giving it a `waiting_time` turns it into a timed gate: agents wait at
the checkpoint for the specified number of seconds before continuing.

Use `set_checkpoint_waiting_time(stage_id, waiting_time)` to update an
existing checkpoint.

```python
from datetime import datetime
print(f"Executed on {datetime.now().strftime('%d.%m.%Y, %H:%M')}")
```

Load a scenario that already contains a checkpoint, then list stages to find the id:

```python
from jupedsim_scenarios import load_scenario

scenario = load_scenario("../assets/waiting-stage-corridor")
scenario.list_stages()
```

Apply the wait by index or by string id:

```python
scenario.set_checkpoint_waiting_time(0, 10.0)                # by index
scenario.set_checkpoint_waiting_time("jps-stages_0", 10.0)   # by string id
```

## Verify the change

Read the underlying stage dict to confirm. The exact key path depends
on the stage type:

```python
import json
print(json.dumps(scenario.stages["jps-stages_0"], indent=2, default=str))
```

## When to use this

Timed checkpoints model real-world holdups:

- **Boarding doors**: a fixed dwell time before agents continue.
- **Security checks**: per-agent processing delay.
- **Phased evacuations**: stagger groups by delaying upstream
  checkpoints.

To explore the effect of different wait times, sweep the value with
`run_sweep` (see *How do I sweep parameters?*).
